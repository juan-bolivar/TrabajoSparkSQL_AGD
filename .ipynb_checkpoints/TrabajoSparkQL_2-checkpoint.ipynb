```python
# 2. Cree dos tablas SparkSQL y almacene el csv en su correspondiente tabla.
# Observación: tenga especial cuidado con los encabezados de los archivos CSV.
# Usted puede considerar cambiar los encabezados de los CSV originales
# Tu código a continuación...

from pyspark.sql.functions import *

df_Asistencia = SesionSpark.read.load("/files/Asistencia.csv",
                          format="csv",
                          sep=",",
                          inferSchema="true",
                          header="true",encoding='UTF-8')


df_Asistencia = df_Asistencia.na.replace('Pertenezco a:','')
df_Asistencia = df_Asistencia.na.replace('Expone:','')

df_Evaluacion = SesionSpark.read.load("/files/Evaluacion.csv",
                          format="csv",
                          sep=",",
                          inferSchema="true",
                          header="true",encoding='UTF-8')

#df_Evaluacion.columns = df_Evaluacion.columns.str.replace(' ', '')



df_Asistencia = df_Asistencia.withColumn('Equipo_al_que_perteneces', regexp_replace("Equipo_al_que_perteneces","Pertenezco a:",""))


df_Asistencia = df_Asistencia.withColumn('Equipo_que_va_a_exponer', regexp_replace("Equipo_que_va_a_exponer","Expone:",""))

df_Asistencia.createOrReplaceTempView('df_Asistencia')

df_Evaluacion.createOrReplaceTempView('df_Evaluacion')
```

```python
# 2. Consulte el listado total de estudiantes (correos electrónicos) del 
# curso de Ingeniería de Software, ordenados alfabéticamente
# Tu código a continuación...


SesionSpark.sql("""
SELECT
Nombre_de_usuario
FROM
    df_Asistencia
ORDER BY Nombre_de_usuario ASC
""").show()

```

```python
# 3. Consulte la cantidad de asistencias registradas por estudiante; además, la fecha y hora de la primera asistencia
# y la fecha y hora de la última asistencia
# Tu código a continuación...


SesionSpark.sql("""
SELECT
Nombre_de_usuario , COUNT(Nombre_de_usuario) AS No_Asistencias , MIN(Marca_Temporal)  AS Primera_Asistencia, MAX(Marca_Temporal) AS Ultima_Asistencia
FROM
    df_Asistencia
    
GROUP BY Nombre_de_usuario 
""").show()
```

```python
# 4. Consulte el listado de estudiantes que asistieron a 2 presentaciones o menos (una).
# Tu código a continuación...



SesionSpark.sql("""
SELECT
Nombre_de_usuario , COUNT(Nombre_de_usuario) AS No_Asistencias 
FROM
    df_Asistencia
    
GROUP BY Nombre_de_usuario 
HAVING No_Asistencias <= 2
""").show()
```

```python
# 5. Consulte el listado de estudiantes que no asistieron a ninguna presentación.
# Tu código a continuación...


SesionSpark.sql("""
SELECT
TABLA1.Nombre_de_usuario   , COUNT(TABLA2.Nombre_de_usuario) AS ASISTENCIAS
FROM
    df_Evaluacion AS TABLA1    
LEFT JOIN df_Asistencia AS TABLA2 ON TABLA1.Nombre_de_usuario = TABLA2.Nombre_de_usuario
GROUP BY TABLA1.Nombre_de_usuario
HAVING ASISTENCIAS = 0

""").show()
```

```python
# 6. Consulte los integrantes por cada equipo al que pertenecen.
# Tu código a continuación...

SesionSpark.sql("""
SELECT
Equipo_al_que_perteneces, collect_list(Nombre_de_usuario) USU
FROM
    df_Asistencia
GROUP BY Equipo_al_que_perteneces
""").show()

```

```python
# 7. Consulte la cantidad de asistentes por presentación, sin considerar los asistentes que pertenecen 
# al equipo que realizó la presentación.
# Tu código a continuación...

SesionSpark.sql("""
SELECT
Equipo_que_va_a_exponer  AS FILTER, SUM(
CASE WHEN Equipo_que_va_a_exponer = Equipo_al_que_perteneces THEN 0 ELSE 1 END
) AS Conteo_Otros_estudiantes
FROM
df_Asistencia
GROUP BY Equipo_que_va_a_exponer
""").show()
```

```python
# 8. Consutar cuáles integrantes evaluaron a su propio equipo. Estas evaluaciones no serán válidas, pues un
# integrante no puede evaluar a su propio equipo.
# Tu código a continuación...

SesionSpark.sql("""
SELECT
TABLA1.Nombre_de_usuario , TABLA2.Equipo_al_que_perteneces , TABLA1.Equipo_que_vas_a_evaluar
FROM
df_Evaluacion AS TABLA1
LEFT JOIN df_Asistencia AS TABLA2 ON TABLA1.Nombre_de_usuario = TABLA2.Nombre_de_usuario
WHERE TABLA2.Equipo_al_que_perteneces = TABLA1.Equipo_que_vas_a_evaluar
""").show()

```
